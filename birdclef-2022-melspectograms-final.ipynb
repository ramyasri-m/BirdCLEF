```python
import os
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

audio_dir = "/kaggle/input/birdclef-2022/train_audio"

def create_mel_spectrogram(audio_path, n_mels=128, fmax=8000, sr=32000, duration=5):
   
    y, sr = librosa.load(audio_path, sr=sr, duration=duration)
    
    mel_spec = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=n_mels, fmax=fmax)

    mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)

    plt.figure(figsize=(10, 4))
    librosa.display.specshow(mel_spec_db, x_axis='time', y_axis='mel', sr=sr)
    plt.colorbar(format='%+2.0f dB')
    plt.title(f'Mel Spectrogram: {os.path.basename(audio_path)}')
    plt.show()
    
    return mel_spec_db
```

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
import os
import gc
import shutil
from concurrent.futures import ProcessPoolExecutor
from tqdm import tqdm
import matplotlib

matplotlib.use('Agg') 

def create_mel_spectrogram_chunked(audio_path, n_mels=128, fmax=8000, sr=32000, duration=5):
    y, sr = librosa.load(audio_path, sr=sr, duration=duration)
    mel_spec = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=n_mels, fmax=fmax)
    mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)
    del y 
    gc.collect()
    
    return mel_spec_db

def save_mel_spectrogram(audio_path, save_dir, n_mels=128, fmax=8000, sr=32000, duration=5):
    mel_spec_db = create_mel_spectrogram_chunked(audio_path, n_mels, fmax, sr, duration)

    audio_filename = os.path.basename(audio_path).replace('.ogg', '.png')
    save_path = os.path.join(save_dir, audio_filename)

    plt.figure(figsize=(10, 4))
    librosa.display.specshow(mel_spec_db, x_axis='time', y_axis='mel', sr=sr)
    plt.colorbar(format='%+2.0f dB')
    plt.title(f'Mel Spectrogram: {audio_filename}')
    plt.savefig(save_path)
    plt.close()
    del mel_spec_db
    gc.collect()

    print(f"Saved Mel spectrogram: {save_path}")

def audio_file_generator(audio_dir, batch_size=5):
    all_audio_files = []
    
    for species in os.listdir(audio_dir):
        species_dir = os.path.join(audio_dir, species)
        if os.path.isdir(species_dir):
            for audio_file in os.listdir(species_dir):
                if audio_file.endswith('.ogg'):
                    audio_path = os.path.join(species_dir, audio_file)
                    all_audio_files.append(audio_path)
    
    print(f"Found {len(all_audio_files)} audio files.") 
    
    # Yield in batches
    for i in range(0, len(all_audio_files), batch_size):
        yield all_audio_files[i:i + batch_size]

def process_and_save_batch(audio_files_batch, save_dir):
    for audio_file in audio_files_batch:
        save_mel_spectrogram(audio_file, save_dir)
    gc.collect()

def process_all_audio_files(audio_dir, save_dir, batch_size=5, n_workers=2):
    os.makedirs(save_dir, exist_ok=True)
    audio_gen = audio_file_generator(audio_dir, batch_size)

    with ProcessPoolExecutor(max_workers=n_workers) as executor:
        futures = []
        for audio_files_batch in audio_gen:
            print(f"Submitting batch of {len(audio_files_batch)} files...")
            futures.append(executor.submit(process_and_save_batch, audio_files_batch, save_dir))
        
        for future in tqdm(futures, desc="Processing batches", unit="batch"):
            future.result()

    gc.collect()
    print("All batches processed. Memory cleaned up.")

audio_dir = "/kaggle/input/birdclef-2022/train_audio"
save_dir = "/kaggle/working/MelSpectograms_Outputs_2"

process_all_audio_files(audio_dir, save_dir, batch_size=5, n_workers=2)
```
